```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Reading, Writing, and Displaying Imagee
img = cv2.imread("D:/OneDrive/Documents/image7.png")  
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
plt.imshow(img)
plt.axis("off")  
plt.show()
```

```python
# 2. Resizing and Cropping
resized_img = cv2.resize(img, (300, 300))
cropped_img = img[50:200, 50:200]
plt.imshow(resized_img)
plt.axis("off")  
plt.show()
```

```python
# Create a green image (500x500 pixels)
G = np.zeros((500, 500, 3), dtype=np.uint8)  # Black image
G[:] = (0, 255, 0)  # Fill with green color (BGR format)
G_rgb = cv2.cvtColor(G, cv2.COLOR_BGR2RGB)

plt.imshow(G_rgb)
plt.axis("off")  # Hide axis
plt.show()
```

```python
plt.imshow(cropped_img)
plt.axis("off")  
plt.show()
```

```python
# 3. Color Spaces
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(gray_img)
plt.axis("off")  
plt.show()
```

```python
plt.imshow(hsv_img)
plt.axis("off")  
plt.show()
```

```python
# 5. Blurring
blurred_img = cv2.GaussianBlur(img, (15, 15), 0)
plt.imshow(blurred_img)
plt.axis("off")  
plt.show()
```

```python
# 6. Edge Detection
edges = cv2.Canny(img, 100, 200)
plt.imshow(edges)
plt.axis("off")  
plt.show()
```

```python
# 7. Contours
# Apply thresholding
_, thresh_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(thresh_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)
# Display the image with contours
plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# 8. Morphological Operations
kernel = np.ones((5,5), np.uint8)
eroded = cv2.erode(thresh_img, kernel, iterations=1)
dilated = cv2.dilate(thresh_img, kernel, iterations=1)
plt.imshow(dilated)
plt.axis("off")  
plt.show()
```

```python
plt.imshow(eroded)
plt.axis("off")  
plt.show()
```

```python
# 9. Feature Detection (ORB)
orb = cv2.ORB_create()
kp = orb.detect(gray_img, None)
kp_img = cv2.drawKeypoints(img, kp, None, color=(0,255,0))
plt.imshow(kp_img)
plt.axis("off")  
plt.show()
```

```python
# 10. Face Detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray_img, 1.1, 4)
face_img = img.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 0, 0), 2)
plt.imshow(face_img)
plt.axis("off")  
plt.show()
```
