```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
def type_cast(data,save_path='encoders.pkl',is_train = True):
    encoders = {}
    for column in data.select_dtypes(include=['bool']).columns:
        data[column] = data[column].astype(int)
    for column in data.select_dtypes(include='object').columns:
        unique_count = data[column].nunique()  
        if unique_count <= 5:
            le = LabelEncoder()
            if is_train:
                data[column] = le.fit_transform(data[column])
                encoders[column] = le
            else:
                if column in encoders:
                    data[column] = encoders[column].transform(data[column])
                else:
                    data[column] = -1
        else:
            if is_train:
                freq_encoding = data[column].value_counts() / len(data)
                data[column] = data[column].map(freq_encoding)
                encoders[column] = freq_encoding
            else:
                if column in encoders:
                    data[column] = data[column].map(encoders[column]).fillna(0)
                else:
                    data[column] = 0
    if is_train:
        joblib.dump(encoders,save_path)
    return data
data = type_cast(data)
```

```python
def detect_outliers(data):
    outliers = {}
    for column in data.columns:
        Q1 = np.percentile(data[column], 25)
        Q3 = np.percentile(data[column], 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        colmn_outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        if not colmn_outliers.empty:
            outliers[column] = colmn_outliers
        if not colmn_outliers.empty:
            print(f"Outliers in column '{column}':")
            print(colmn_outliers[column].values)
            print("="*50)

    return outliers
outliers = detect_outliers(data)       
```

```python
def cap_outliers(df, lower_percentile=25, upper_percentile=75):
    for column in df.select_dtypes(include='number').columns:
        # Calculate the lower and upper bounds
        lower_bound = np.percentile(df[column], lower_percentile)
        upper_bound = np.percentile(df[column], upper_percentile)
        
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    
    return df

df_capped = cap_outliers(data)
df_capped
```
