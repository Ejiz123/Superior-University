```python
import pandas as pd
```

```python
def fill_values(data,colmns):
    for colmn in colmns:
        if data[colmn].dtype == object:
            data[colmn] = data[colmn].fillna(data[colmn].mode()[0])
        elif data[colmn].dtype == 'float64':
            data[colmn] = data[colmn].fillna(data[colmn].mean())
        elif data[colmn].dtype == 'int64':
            data[colmn] = data[colmn].fillna(data[colmn].mean())
colmns = ['Revenue (Millions)','Metascore']
fill_values(data,colmns)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def typecast(data,colmns):
    for colmn in colmns:
        if data[colmn].dtype == 'float64':
            data[colmn] = data[colmn].astype('int64')
        elif data[colmn].dtype == object:
            labelencoder = LabelEncoder()
            data[colmn] = labelencoder.fit_transform(data[colmn])
            data[colmn] = data[colmn].astype('int64')
colmns = ['Title','Genre','Description','Director','Actors','Rating','Revenue (Millions)','Metascore']
typecast(data,colmns)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(['Title','Description'],axis = 1)
y = data['Revenue (Millions)']
```

```python
X_train,X_test,Y_Train,Y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
categorical_colmns = ['Genre', 'Director', 'Actors']
label_encoders = {}
for col in categorical_colmns:
    label_encoders[col] = LabelEncoder()
    x[col] = label_encoders[col].fit_transform(x[col])
X_train,X_test,Y_Train,Y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_Train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(Y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
```
