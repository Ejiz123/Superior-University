```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Define the trained model (assuming it's already trained)
linear_model = LinearRegression()
# You need to fit the model with training data first.
# Example: linear_model.fit(x_train, y_train)

# Preprocessing function (similar to your previous transformations)
def preprocess_input(region, soil_type, crop, weather_condition, rainfall, temperature, fertilizer_used, irrigation_used, days_to_harvest):
    # Encode categorical features using LabelEncoder
    le = LabelEncoder()
    
    # Example labels for encoding
    region_labels = ['North', 'East', 'South', 'West']
    soil_type_labels = ['Clay', 'Sandy', 'Loam', 'Silt', 'Peaty', 'Chalky']
    crop_labels = ['Wheat', 'Rice', 'Maize', 'Barley', 'Soybean', 'Cotton']
    weather_condition_labels = ['Sunny', 'Rainy', 'Cloudy']
    
    # Encoding categorical data
    region_encoded = le.fit_transform(region_labels).tolist()
    soil_type_encoded = le.fit_transform(soil_type_labels).tolist()
    crop_encoded = le.fit_transform(crop_labels).tolist()
    weather_condition_encoded = le.fit_transform(weather_condition_labels).tolist()
    
    # Prepare the input array in the same order as your model's training data
    input_data = np.array([[region_encoded, soil_type_encoded, crop_encoded, weather_condition_encoded,
                            rainfall, temperature, fertilizer_used, irrigation_used, days_to_harvest]])
    
    return input_data

# Function to predict crop yield based on user input
def predict_yield():
    # Taking user inputs for features
    region = input("Enter region (North, East, South, West): ")
    soil_type = input("Enter soil type (Clay, Sandy, Loam, Silt, Peaty, Chalky): ")
    crop = input("Enter crop (Wheat, Rice, Maize, Barley, Soybean, Cotton): ")
    weather_condition = input("Enter weather condition (Sunny, Rainy, Cloudy): ")
    rainfall = float(input("Enter rainfall (in mm): "))
    temperature = float(input("Enter temperature (in Celsius): "))
    fertilizer_used = input("Was fertilizer used? (True/False): ").lower() == 'true'
    irrigation_used = input("Was irrigation used? (True/False): ").lower() == 'true'
    days_to_harvest = int(input("Enter the number of days to harvest: "))
    
    # Preprocess input data
    input_data = preprocess_input(region, soil_type, crop, weather_condition, rainfall, temperature, fertilizer_used, irrigation_used, days_to_harvest)
    
    # Make prediction
    predicted_yield = linear_model.predict(input_data)
    
    print(f"Predicted Crop Yield: {predicted_yield[0]:.2f} tons per hectare")

# Call the function
predict_yield()
```
